```python
import csv
import pandas as pd 
import numpy as np 
import requests
import json 
import time 

#bring in the list of adjectives and verbs 
adjectives = pd.read_csv('C:/Users/nsw26/Desktop/Coding projects/domainsearch/adjectives.csv')
verbs = pd.read_csv('C:/Users/nsw26/Desktop/Coding projects/domainsearch/verbs.csv')

#create the dataFrame with the domains formed via combinations of verbs and adjectives 
organized=pd.DataFrame()
organized['Adjectives']=adjectives
organized['Verbs']=verbs 
organized['Combined']=organized['Verbs']+organized['Adjectives']+".com"

#organized.reset_index(drop=True, inplace=True) 

combinedList=organized['Combined']

#create the scoring system 
combinedLength = []
for x in combinedList: 
    length=len(x)
    score=20-length
    combinedLength.append(score)
    
organized['Score'] = combinedLength

#arrange by score 
sub=organized[['Combined','Score']].sort_values(by=['Score'],ascending=False)

print(sub)

topDomains=sub.iloc[:49]
topDomains=topDomains['Combined']



```
